```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
file="../data/fan.csv"
df = pd.read_csv(file)
df.head()
```

```python
pd.set_option("display.max.columns", None)
```

```python
healthy=df["Faulty"]==0
```

```python
healthy=df[healthy].head(100)
```

```python
faulty=df["Faulty"]==1
```

```python
faulty=df[faulty].head(100)
```

```python
fig = plt.figure()
ax1 = fig.add_subplot(111)
ax1.scatter(healthy["Time"],healthy["Sound"],label='Healthy')
ax1.scatter(faulty["Time"],faulty["Sound"],label='Faulty')
frame1 = plt.gca()
frame1.axes.get_xaxis().set_visible(False)
plt.legend(loc='upper left');
plt.title("Sound")
```

```python
fig = plt.figure()
ax1 = fig.add_subplot(111)
ax1.scatter(healthy["Time"],healthy["Vibration"],label='Healthy')
ax1.scatter(faulty["Time"],faulty["Vibration"],label='Faulty')
frame1 = plt.gca()
frame1.axes.get_xaxis().set_visible(False)
plt.legend(loc='upper left');
plt.title("Vibration")
```
